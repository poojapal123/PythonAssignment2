<b> k. Line chart to show the trend of data - All Numeric/Date Variables

```python
# change the invoice_date format from String to Timestamp

df['InvoiceDate'] = pd.to_datetime(df.InvoiceDate, format='%m/%d/%Y %H:%M')
```

```python
df.insert(loc=3, column='Year_month', value=df['InvoiceDate'].map(lambda x: 100*x.year + x.month))
df.insert(loc=4, column='Month',value=df.InvoiceDate.dt.month)
df.insert(loc=5, column='Day',  value=df.InvoiceDate.dt.dayofweek) 
df.insert(loc=6, column='Hour', value=df.InvoiceDate.dt.hour)
```

```python
df['Month'] = df['InvoiceDate'].apply(lambda x: pd.Timestamp(x).strftime('%B'))
df['Day'] = df['InvoiceDate'].apply(lambda x: pd.Timestamp(x).strftime('%A'))
```

```python
df['Revenue']=df['Quantity']*df['UnitPrice']
```

```python
df.head(2)
```

```python
plt.subplots(figsize=(15,5))
sb.set()
sb.lineplot(data=df, x='Month',y='Revenue');
plt.title('LINE CHART TO SHOW THE TREND OF DATA');
```

<b> 2. Check for missing values in all columns and replace them with the appropriate metric 
(Mean/Median/Mode) 

```python
df.isnull().sum()
```

only description and CustomerID contains missing value 

As we earlier plotted data distribution graph of CustomerID so based on the same graph we analyze that data distribution of CustomerID is symmetric the appropriate metric is mean 
So, we have to replace missing values with mean



```python
df['CustomerID'].fillna(df['CustomerID'].mean(),inplace=True)
```

```python
df['Description'].fillna(df['Description'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```
